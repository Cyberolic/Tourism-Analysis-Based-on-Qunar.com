```python
csv_1=open('Data.csv','w',encoding='utf-8-sig',newline='')
csv_write=csv.writer(csv_1)
csv_write.writerow(['旅游地点','标题','出发时间','天数','人均费用','人物','玩法','浏览量','详情URL'])



for page in range(1,200):
    print('正在读取页数为:',str(page))
    #伪造表头(可以省略)
    user=["Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1",
                "Mozilla/5.0 (X11; CrOS i686 2268.111.0) AppleWebKit/536.11 (KHTML, like Gecko) Chrome/20.0.1132.57 Safari/536.11",
                "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/536.6 (KHTML, like Gecko) Chrome/20.0.1092.0 Safari/536.6",
                "Mozilla/5.0 (Windows NT 6.2) AppleWebKit/536.6 (KHTML, like Gecko) Chrome/20.0.1090.0 Safari/536.6",
                "Mozilla/5.0 (Windows NT 6.2; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/19.77.34.5 Safari/537.1",
                "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/536.5 (KHTML, like Gecko) Chrome/19.0.1084.9 Safari/536.5",
                "Mozilla/5.0 (Windows NT 6.0) AppleWebKit/536.5 (KHTML, like Gecko) Chrome/19.0.1084.36 Safari/536.5",
                "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/536.3 (KHTML, like Gecko) Chrome/19.0.1063.0 Safari/536.3",
                "Mozilla/5.0 (Windows NT 5.1) AppleWebKit/536.3 (KHTML, like Gecko) Chrome/19.0.1063.0 Safari/536.3",
                "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_8_0) AppleWebKit/536.3 (KHTML, like Gecko) Chrome/19.0.1063.0 Safari/536.3",
                "Mozilla/5.0 (Windows NT 6.2) AppleWebKit/536.3 (KHTML, like Gecko) Chrome/19.0.1062.0 Safari/536.3",
                "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/536.3 (KHTML, like Gecko) Chrome/19.0.1062.0 Safari/536.3",
                "Mozilla/5.0 (Windows NT 6.2) AppleWebKit/536.3 (KHTML, like Gecko) Chrome/19.0.1061.1 Safari/536.3",
                "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/536.3 (KHTML, like Gecko) Chrome/19.0.1061.1 Safari/536.3",
                "Mozilla/5.0 (compatible; MSIE 9.0; Windows NT 6.1; Trident/5.0",
                "Mozilla/4.0 (compatible; MSIE 8.0; Windows NT 6.0; Trident/4.0)",
                "Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 6.0)",
                "Mozilla/4.0 (compatible; MSIE 6.0; Windows NT 5.1)"
    ]
    user_agent = random.choice(user)
    headers={'user-agent': user_agent}
    #发送请求
    url='https://travel.qunar.com/travelbook/list.htm?page={}&order=hot_heat'.format(page)
    response=requests.get(url,headers=headers)
    #获取原html
    data_html=response.text
    selector=parsel.Selector(data_html)
    #提取每个用户的标签
    url_list=selector.css('body > div.qn_mainbox> div > div.left_bar > ul > li > h2 > a::attr(href)').getall()
    #拿到list向每一个详情页发出请求(样板如下) 
    for url in url_list:
        try:
            id=url.replace('/youji/','')
            #拼接连接
            url_x='https://travel.qunar.com/travelbook/note/'+id
            print(url_x)
            #建立连接获取数据和网页源代码
            time.sleep(random.randint(5,20))
            res_1=requests.get(url_x,headers=headers)
            data_html_1=res_1.text
            #提取数据，获取里面的文本内容
            selector_1=parsel.Selector(data_html_1)
            #获取旅游城市
            title=selector_1.css('.b_crumb_cont *:nth-child(3)::text').get()
            #获得标题
            comment=selector_1.css('#booktitle::text').get()
            #获得日期
            date=selector_1.css('#js_mainleft > div.b_foreword > ul > li.f_item.when > p > span.data::text').get()
            #天数
            howlong=selector_1.css('#js_mainleft > div.b_foreword > ul > li.f_item.howlong > p > span.data::text').get()
            #人均费用
            avg_exp=selector_1.css('#js_mainleft > div.b_foreword > ul > li.f_item.howmuch > p > span.data::text').get()
            #人物
            who=selector_1.css('#js_mainleft > div.b_foreword > ul > li.f_item.who > p > span.data::text').get()
            #玩法
            ways_to_play=selector_1.css('#js_mainleft > div.b_foreword > ul > li.f_item.how > p >span.data span::text').get()
            li=list(ways_to_play)
            play=''.join(li)
            #浏览量
            num=selector_1.css('.view_count::text').get()
            print(title,comment,date,howlong,avg_exp,who,play,num,url_x)
            
            #写入数据
            a=[title,comment,date,howlong,avg_exp,who,play,num,url_x]
            csv_write.writerow(a)
    
        except TypeError:
            continue
            
    

            
csv_1.close()
        
        

    
    
```
